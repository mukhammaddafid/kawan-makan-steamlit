## Filter data

### Subtask:
Filter the DataFrame to include only rows for 'Maize and products', 'Rice (Milled Equivalent)', and 'Wheat and products' with the unit '1000 tonnes'.


**Reasoning**:
Filter the DataFrame to include only the specified staple crops and unit.



```python
staple_crops_list = ['Maize and products', 'Rice (Milled Equivalent)', 'Wheat and products']
df_staple_filtered = df[df['Item'].isin(staple_crops_list) & (df['Unit'] == '1000 tonnes')]
display(df_staple_filtered.head())
```

**Reasoning**:
Identify the year columns and melt the DataFrame to a long format, then display the head of the melted DataFrame to verify the transformation.



```python
year_cols = [col for col in df_staple_filtered.columns if col.startswith('Y')]
df_melted = pd.melt(df_staple_filtered,
                    id_vars=['Area', 'Item', 'Element'],
                    value_vars=year_cols,
                    var_name='Year',
                    value_name='Value')
display(df_melted.head())
```

## Aggregate data

### Subtask:
Group the reshaped data by 'Year' and 'Item' and sum the production values to get the total production for each item in each year across all areas.


```python
df_aggregated_production = df_melted.groupby(['Year', 'Item'])['Value'].sum().reset_index()
display(df_aggregated_production.head())
```

```python
fig = px.line(df_aggregated_production, x='Year', y='Value', color='Item',
              title='Production Evolution of Maize, Rice, and Wheat (1000 tonnes)')
fig.update_layout(xaxis_title='Year', yaxis_title='Production (1000 tonnes)')
fig.show()
```
